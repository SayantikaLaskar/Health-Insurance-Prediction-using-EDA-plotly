```python
# Formatting insurance company names
x=x.replace({"Health":"","General":"","Co. Ltd.":"", "Insurance":""}, regex=True) #Removing the words: Insurance, Co. Ltd. from the insurance company names
x.insurer = x.insurer.str.strip() #Removing leading and trailing spaces
```

```python
# Adding a column that specifies the type of undertaking
df = x.copy()
undertaking_dict = {# Storing the insurer names for each undertaking
    "public"    :df.loc[df.index<=4]['insurer'].to_list(),
    "private"   :df.loc[(df.index>=7) & (df.index<=28)]['insurer'].to_list(),
    "standalone":df.loc[(df.index>=31) & (df.index<=37)]['insurer'].to_list(),
}

def fill_undertaking(df, undertaking_dict):
    """
    Creates and fills the undertaking column of a dataframe based on the insurer names

        Parameters:
            df(pd.DataFrame): The dataframe that needs the undertaking column
            undertaking_dict(dictionary): Pass the same variable

        Returns:
            df(pd.DataFrame): altered dataframe
            undertaking_names(array): undertaking values
    """
    df['undertaking'] = df['insurer']
    for key, value in list(undertaking_dict.items()):
        df['undertaking'] = df['undertaking'].replace(dict.fromkeys(value, key))

    df.loc[df.insurer.str.contains("Total", regex = True), "undertaking"] = 'aggregate'
    undertaking_names = x.undertaking.values

    return df, undertaking_names

x, undertaking_names = fill_undertaking(x, undertaking_dict)
```

```python
# Iterating through the years and storing the total column value for every scheme in a dictionary

persons_covered_dict, policy_count_dict, gross_premium_dict = {}, {}, {} #Contains the respective sub-levels of totals colums for each year

for year in fin_years:

    try:
        persons_covered_dict[year] = x[year]["total"]["persons_covered"]
        policy_count_dict[year] = x[year]["total"]["policy_count"]
        gross_premium_dict[year] = x[year]["total"]["gross_premium"]
    except:
        print(f"Error in year => [{year}]")

total_dict = {"persons_covered_dict":persons_covered_dict,
              "policy_count_dict":policy_count_dict,
              "gross_premium_dict":gross_premium_dict}

insurer_names = x.insurer.values
```

Comparing the average premium per policy for different schemes

```python
import pandas as pd
import numpy as np

def df_formatter_for_avg_prem(scheme, master_df, fin_years, undertaking_dict):
    """
    Function takes in the original large dataframe and returns a melted dataframe for the scheme entered.
    The melted dataframe contains the insurers and the average premium per person for all the years.

    Parameters:
        scheme(str): The scheme for which you want the function to extract the data
        master_df(pd.DataFrame): The master dataframe from which the extraction should take place
        fin_years, undertaking_dict: Pass the same variables
    Returns:
        df_long(pd.DataFrame): Altered long dataframe
        df_wide(pd.DataFrame): Altered wide dataframe
    """
    print(f"==>Generating melted dataframe for scheme: {scheme}")

    # Assuming that 'insurer_names' is defined somewhere before this function is called
    df_wide = pd.DataFrame(columns=["insurer", *fin_years])
    df_wide["insurer"] = insurer_names  # You need to define 'insurer_names' somewhere

    for year in fin_years:
        # Assuming that 'x' is defined somewhere before this function is called
        df = x[year].loc[:, [(scheme, "persons_covered"), (scheme, "gross_premium")]].copy()
        df = df.droplevel(0, axis=1)
        df["avg_prem_per_policy"] = np.round(df["gross_premium"] / df["persons_covered"], 0)
        df_wide[year] = df["avg_prem_per_policy"]
        # except:
        #     print(f"{scheme} does not exist in {year}")

    df_long = pd.melt(
        df_wide,
        id_vars="insurer",
        var_name="year",
        value_name="avg_prem_per_policy"
    )

    df_long, _ = fill_undertaking(df_long, undertaking_dict)

    # df_long = df_long.style.set_caption(f'Dataframe to plot average premium per policy for the scheme: {scheme}')

    print(f"---{scheme} df complete---")

    return df_long, df_wide
```
